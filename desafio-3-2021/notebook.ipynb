```python
from mlxtend.frequent_patterns import apriori

df_merged.replace(to_replace='T', value=True, inplace=True)
df_merged.replace(to_replace='F', value=False, inplace=True)

df_merged_temp = df_merged.drop(columns=['Idade', 'Renda', 'Regiao', 'Genero'])
frequent_itemsets = apriori(df_merged_temp, min_support=0.1, max_len=6, use_colnames=True)


df_merged_temp_hombres = df_merged[df_merged['Genero'] == 1].drop(columns=['Idade', 'Renda', 'Regiao', 'Genero'])
df_merged_temp_mujeres = df_merged[df_merged['Genero'] == 0].drop(columns=['Idade', 'Renda', 'Regiao', 'Genero'])

frequent_itemsets_hombres = apriori(df_merged_temp_hombres, min_support=0.1, max_len=6, use_colnames=True)

frequent_itemsets_mujeres = apriori(df_merged_temp_mujeres, min_support=0.1, max_len=6, use_colnames=True)

regiaos = ["CO", "SE", "NE", "S", "N"]
rules_by_region = {}
for region in regiaos:
  aux = df_merged.drop(columns=['Idade', 'Renda', 'Genero'])
  aux = aux[ aux["Regiao"] == region ]
  frequent_itemsets_aux = apriori(aux.drop(columns=['Regiao']), min_support=0.1, max_len=6, use_colnames=True)
  rules_by_region[region] = association_rules(frequent_itemsets_aux, metric="confidence", min_threshold=0.8)

```

```python
from mlxtend.frequent_patterns import association_rules


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)

rules_hombres = association_rules(frequent_itemsets_hombres, metric="confidence", min_threshold=0.8)

rules_mujeres = association_rules(frequent_itemsets_mujeres, metric="confidence", min_threshold=0.8)
```

```python
from collections import Counter

products = [
  "seguro auto",
  "seguro vida Emp",
  "seguro vida PF",
  "Seguro Residencial",
  "Investimento Fundos_cambiais",
  "Investimento Fundos_commodities",
  "Investimento LCI",
  "Investimento LCA",
  "Investimento Poupanca",
  "Investimento Fundos Multimercado",
  "Investimento Tesouro Direto",
  "Financiamento Casa",
  "Financiamento Carro",
  "Emprestimo _pessoal",
  "Emprestimo _consignado",
  "Emprestimo _limite_especial",
  "Emprestimo _educacao",
  "Emprestimo _viagem",
  "Investimento CDB",
  "Investimento Fundos"
]

def get_products_contratados(transaction):
  productos_contratados = [product for product in products if transaction[product]]
  return productos_contratados


def get_rules_for(productos_contratados, rules):
  rule_to_aply = []
  for index, rule in rules.iterrows():
    check =  all(item in productos_contratados for item in rule['antecedents'])
    #if Counter(rule['antecedents']) == Counter(productos_contratados):
    if check:
      rule_to_aply.append(rule)
  df = pd.DataFrame(rule_to_aply)
  if len(rule_to_aply) > 0:
    return df.sort_values(by=['confidence'], ascending=False).head()
  else:
    return df


def get_sugerencias(rules_to_aply, productos_contratados):
  sugerencias = {}
  for index, rule in rules_to_aply.iterrows():
    for sugerencia in rule['consequents']:
      if sugerencia not in sugerencias and sugerencia not in productos_contratados:
        sugerencias[sugerencia] = rule['confidence']
      if len(sugerencias) >= 3:
        break
    if len(sugerencias) >= 3:
      break

  return sugerencias
```
